# Melt()

```python
import pandas as pd
```

```python
var = pd.DataFrame({"days":[1,2,3,4,5],"eng":[10,12,13,14,15],"Math":[17,18,19,20,21]})
var
```

# Reshape

```python
pd.melt(var)
```

# change index name

```python
pd.melt(var,id_vars=["eng"])
```

# Pivot

```python
var = pd.DataFrame({"days":[1,2,3,4,5],"st_name":["a","b","c","d","e"], "eng":[10,12,13,14,15],"Math":[17,18,19,20,21]})
var
```

```python
var.pivot(index="days",columns="st_name")
```

```python
var.pivot(index="days",columns="st_name",values="eng")
```

# MEAN1

```python
 var.pivot_table(index="st_name",columns="days",aggfunc="mean")
```

# SUM

```python
var.pivot_table(index="st_name",columns="days",aggfunc="sum")
```

# Margins

```python
var.pivot_table(index="st_name",columns="days",aggfunc="sum",margins=True)
```
